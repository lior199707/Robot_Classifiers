# Imports

```python
import pandas as pd
import joblib
from sklearn.preprocessing import LabelEncoder
import numpy as np
from sklearn.metrics import accuracy_score, classification_report
```

# Load best models

```python
# Load best rocket model
best_rocket_model = joblib.load("./eda_rocket/model/eda_rocket_model.pkl")
```

```python
# Load best weasel model
best_weasel_model = joblib.load("./eda_weasel/model/eda_weasel_model1.pkl")
```

```python
# Load best shapelet model
best_shapelet_model = joblib.load("./eda_shapelet/model/eda_shapelet_model1.pkl")
```

# Organaize data for testing

```python
test_data = pd.read_csv("eda_rocket/train_data/EDA_vlad_1_results.csv")
# Take segments id 19 to 58
test_data = test_data[380:1180]
test_data
```

```python
Y_test = pd.read_csv("eda_rocket/train_data/EDA_vlad_1.csv")
# Take segments id 19 to 58
Y_test = Y_test[19:59]
Y_test
```

```python
label_encoder = LabelEncoder()
# Get the real labels of the testing data
test_encoded_labels = label_encoder.fit_transform(Y_test.label)
test_encoded_labels
```

```python
Y_test["encoded_label"] = test_encoded_labels
Y_test
```

```python
# Extract the measuremens columns as time series data
X_test = test_data.groupby('Segment id')['EdaArousalMean []'].apply(list).values
X_test_np = np.vstack(X_test)
X_test_np
```

# Test models on Vlad's 1st experiment data

## Test best rocket model

```python
# Print the real labels
test_encoded_labels
```

```python
y_pred_rocket = best_rocket_model.predict(X_test_np)
y_pred_rocket
```

```python
# Calculate accuracy of the model on Vlad's 1st experiment data
accuracy_rocket = accuracy_score(test_encoded_labels, y_pred_rocket)
print(f"Accuracy: {accuracy_rocket * 100:.2f}%")
report_rocket = classification_report(test_encoded_labels, y_pred_rocket)
print(report_rocket)
```

## Test best weasel model

```python
y_pred_weasel = best_weasel_model.predict(X_test_np)
y_pred_weasel
```

```python
# Calculate accuracy of the model on Vlad's 1st experiment data
accuracy_weasel = accuracy_score(test_encoded_labels, y_pred_weasel)
print(f"Accuracy: {accuracy_weasel * 100:.2f}%")
report_weasel = classification_report(test_encoded_labels, y_pred_weasel, zero_division=1)
print(report_weasel)
```

## Test best shapelet model

```python
y_pred_shapelet = best_shapelet_model.predict(X_test_np)
y_pred_shapelet
```

```python
# Calculate accuracy of the model on Vlad's 1st experiment data
accuracy_shapelet = accuracy_score(test_encoded_labels, y_pred_shapelet)
print(f"Accuracy: {accuracy_shapelet * 100:.2f}%")
report_shapelet = classification_report(test_encoded_labels, y_pred_shapelet)
print(report_shapelet)
```
